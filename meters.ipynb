```python
window_size = 2 * 5
```

```python
profile = mp.compute(sub['active_power_per_device'].to_numpy(), windows=window_size)
figures = mp.visualize(profile)
ax = plt.gca()
ax.set_title('Matrix profile substation, 1 day window')
```

```python
fig, axs = plt.subplots(max((n + 1) // 2, 2), 2, figsize=(14, 4))
plt.tight_layout()
for i in range(n):
    axs[i // 2, i % 2].plot(lv[lv['lv_id'] == i]['date'].unique(), lv[lv['lv_id'] == i].groupby('date')['active_power_per_device'].mean())
    axs[i // 2, i % 2].set_title(f'LV {i} active power mean by day')
plt.show()
```

```python
plt.tight_layout()
for i in range(n):
    profile = mp.compute(lv[lv['lv_id'] == i]['active_power_per_device'].to_numpy(), windows=window_size)
    figures = mp.visualize(profile)
    ax = plt.gca()
    ax.set_title(f'Matrix profile LV {i}, 1 day window')
```

```python
data = np.zeros((n, 1488))
for i in range(n):
    row = lv[lv['lv_id'] == i]['active_power_per_device'].to_numpy()
    row = (row - row.mean()) / row.std()
    data[i] = row
data
```

```python
for i in range(n):
    plt.plot(meters.mean_profile(data, i, window_size), label=f'LV {i}')
plt.gcf().set_figwidth(15)
plt.gcf().set_figheight(10)
plt.xticks(rotation=90)
plt.legend(labels=[str(i) for i in range(n)], loc='upper right')
```

```python
profile = meters.mean_profile(data, 3, window_size)
```

```python
anomaly_start = np.argmax(profile)
anomaly_start
```

```python
fig, axs = plt.subplots(max((n + 1) // 2, 2), 2, figsize=(14, 4))
plt.tight_layout()
for i in range(n):
    axs[i // 2, i % 2].plot(lv[lv['lv_id'] == i]['active_power_per_device'].to_numpy()[anomaly_start:anomaly_start + window_size + 1])
    axs[i // 2, i % 2].set_title(f'LV {i}')
    axs[i // 2, i % 2].set_ylim(50, 600)
plt.show()
```
